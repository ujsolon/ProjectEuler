```python
def chain(x):
    c=[x]
    xold=x
    xnew=sumdigfac(x)
    while xnew not in c:
        c.append(xnew)
        xold=xnew
        xnew=sumdigfac(xnew)
    return c

def listdiff(fromx,y):
    for item in y:
        if item in fromx:
            fromx.remove(item)
    return fromx

q=[1,2,3,5]
w=[2,5,6]
listdiff(q,w)
```
